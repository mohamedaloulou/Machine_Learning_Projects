```python
#4
# Create a new column "safe_loans" with +1 for safe loans and -1 for risky loans
df['safe_loans'] = df['bad_loans'].apply(lambda x: 1 if x == 0 else -1)
df.drop('bad_loans', axis=1, inplace=True)
print(df['safe_loans'])
```

```python
#5
safe_loan_percentage = len(df[df['safe_loans']==1])
bad_loan_percentage = len(df[df['safe_loans']== -1])
print('number of safe loans are : ',safe_loan_percentage)
print('number of bad loans are : ',bad_loan_percentage )
```

```python
#7
# List of columns to select
selected_columns = [
    'grade', 'sub_grade', 'short_emp', 'emp_length_num', 'home_ownership',
    'dti', 'purpose', 'term', 'last_delinq_none', 'last_major_derog_none',
    'revol_util', 'total_rec_late_fee'
]

# Create a new DataFrame X containing the selected columns
x = df[selected_columns]
y=df['safe_loans']
```

```python
#9,10
safe_loans_raw = df[df['safe_loans']==1]
risky_loans_raw =df[df['safe_loans']== -1]
print('number of safe loans are : ',len(safe_loans_raw))
print('number of bad loans are : ',len(risky_loans_raw) )
print('percentage of safe loans are : ',len(safe_loans_raw)/len(df['safe_loans'])*100)
print('percentage of bad loans are : ',len(risky_loans_raw)/len(df['safe_loans'])*100 )
```

```python
#11
from sklearn.utils import resample

minority_class_size = len(risky_loans_raw)

safe_loans_undersampled = resample(safe_loans_raw, n_samples=minority_class_size, random_state=0)

balanced_dataset = pd.concat([safe_loans_undersampled, risky_loans_raw])
balanced_dataset
```

```python
#12
safe_loans_raw_balanced = balanced_dataset[balanced_dataset['safe_loans']==1]
risky_loans_raw_balanced =balanced_dataset[balanced_dataset['safe_loans']== -1]
print('number of safe loans are : ',len(safe_loans_raw_balanced))
print('number of bad loans are : ',len(risky_loans_raw_balanced) )
print('percentage of safe loans are : ',len(safe_loans_raw_balanced)/len(balanced_dataset['safe_loans'])*100)
print('percentage of bad loans are : ',len(risky_loans_raw_balanced)/len(balanced_dataset['safe_loans'])*100 )
```

```python
# 13
selected_features = ['grade', 'sub_grade', 'short_emp', 'emp_length_num', 'home_ownership', 'dti', 'purpose', 'term', 'last_delinq_none', 'last_major_derog_none', 'revol_util', 'total_rec_late_fee','safe_loans']

bd = balanced_dataset[selected_features]
```

```python
bd= pd.get_dummies(bd)
```

```python
# 14
from sklearn.model_selection import train_test_split

X = bd.drop('safe_loans', axis=1)
y = bd['safe_loans']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 15
from sklearn.tree import DecisionTreeClassifier

decision_tree_model = DecisionTreeClassifier()

decision_tree_model.fit(X_train, y_train)
```

```python
# 16
small_model = DecisionTreeClassifier(max_depth=2, random_state=0)

small_model.fit(X_train, y_train)
```

```python
# 17
import graphviz
from sklearn import tree

def draw_tree(tree_model, features):
    tree_data = tree.export_graphviz(tree_model,
                                     impurity=False,
                                     feature_names=features,
                                     class_names=tree_model.classes_.astype(str),
                                     filled=True,
                                     out_file=None)
    graph = graphviz.Source(tree_data)
    display(graph)
draw_tree(small_model, X_train.columns)
```

```python
#18
predictions = decision_tree_model.predict(X_test)
predictions_proba = decision_tree_model.predict_proba(X_test)
```

```python
# 19
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

accuracy = accuracy_score(y_test, predictions)

f1 = f1_score(y_test, predictions)

conf_matrix = confusion_matrix(y_test, predictions)

print("Accuracy:", accuracy)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
#20
max_depth_values = range(1, 21)
accuracies = []
for max_depth in max_depth_values:
    model = DecisionTreeClassifier(max_depth=max_depth, random_state=0)

    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)
    accuracies.append(accuracy)

best_max_depth = max_depth_values[accuracies.index(max(accuracies))]

for depth, accuracy in zip(max_depth_values, accuracies):
    print(f"Max Depth: {depth}, Accuracy: {accuracy:.4f}")

print("Best Max Depth:", best_max_depth)
```

```python
#21
from sklearn.linear_model import LogisticRegression

logistic_regression_model = LogisticRegression()

logistic_regression_model.fit(X_train, y_train)
```

```python
#22
logistic_regression_predictions = logistic_regression_model.predict(X_test)

logistic_regression_accuracy = accuracy_score(y_test, logistic_regression_predictions)

logistic_regression_f1 = f1_score(y_test, logistic_regression_predictions)

logistic_regression_conf_matrix = confusion_matrix(y_test, logistic_regression_predictions)

print("Logistic Regression Model Evaluation:")
print("Accuracy:", logistic_regression_accuracy)
print("F1 Score:", logistic_regression_f1)
print("Confusion Matrix:")
print(logistic_regression_conf_matrix)
```
